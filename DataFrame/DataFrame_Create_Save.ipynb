<h1>Import Pandas</h1>

```python
import pandas as pd
```

<h1>Create DataFrame using Python Dictionary</h1>

```python
weather_data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017'],
    'temparature': [32, 35, 28, 24, 32, 31],
    'windspeed': [6, 7, 2, 7, 4, 2],
    'event': ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny']
}

df = pd.DataFrame(weather_data)
df
```

<h1>Create DataFrame using CSV File</h1>

```python
df = pd.read_csv('weather_data.csv')
df
```

<h1>Create DataFrame using Tuple List</h1>

```python
weather_data = [
    ('1/1/2017', 32, 6, 'Rain'),
    ('1/2/2017', 35, 7, 'Sunny'),
    ('1/3/2017', 28, 2, 'Snow'),
    ('1/4/2017', 24, 7, 'Snow'),
    ('1/5/2017', 32, 4, 'Rain'),
    ('1/6/2017', 31, 2, 'Sunny')
]

df = pd.DataFrame(weather_data, columns=['day', 'temparature', 'windspeed', 'event'])
df
```

<h1>Create DataFrame using List of Dictionaries</h1>

```python
weather_data = [
    {'day': '1/1/2017', 'temparature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temparature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temparature': 28, 'windspeed': 2, 'event': 'Snow'},
    {'day': '1/4/2017', 'temparature': 24, 'windspeed': 7, 'event': 'Snow'},
    {'day': '1/5/2017', 'temparature': 32, 'windspeed': 4, 'event': 'Rain'},
    {'day': '1/6/2017', 'temparature': 32, 'windspeed': 2, 'event': 'Sunny'}
]

df = pd.DataFrame(weather_data)
df
```
