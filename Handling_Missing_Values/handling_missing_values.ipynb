<h1>parse_dates</h1>

```python
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df
```

```python
type(df.day[0])
```

<h1>set_index</h1>

```python
df.set_index('day', inplace=True)
df
```

<h1>fillna</h1>

<h3>Replace with 0</h3>

```python
new_df = df.fillna(0)
new_df
```

<h3>Replace with Dictionary</h3>

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

<h3>Replace with Forward Fill</h3>

```python
new_df = df.fillna(method='ffill')
new_df
```

<h3>Replace with Backward Fill</h3>

```python
new_df = df.fillna(method='bfill')
new_df
```

<h1>Interpolation</h1>
<p>Default: <b>Linear</b></p>

```python
new_df = df.interpolate()
new_df
```

<p><b>Time</b> Interpolation</p>

```python
new_df = df.interpolate(method='time')
new_df
```

<h1>dropna</h1>

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

<h1>reindex</h1>

```python
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
new_df = df.reindex(dt)
new_df
```

<h1>Relpace</h1>

```python
import numpy as np
```

```python
df = pd.read_csv('weather_data_replace.csv')
df
```

<h3>Using List</h3>

```python
new_df = df.replace(['-99999', '-88888'], np.NaN)
new_df
```

<h3>Using Dictionary with Columns</h3>

```python
new_df = df.replace({
    'temperature': '-99999',
    'windspeed': '-88888',
    'event': '0'
}, np.NaN)

new_df
```

<h3>Using Dictionary with Values</h3>

```python
new_df = df.replace({
    '-99999': np.NaN,
    '-88888': np.NaN,
    '0': 'No Event'
})

new_df
```

<h3>Using RegEx</h3>

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]',
}, '', regex=True)

new_df
```

<h3>Using two list to replace some values</h3>

```python
df = pd.DataFrame({
    'score': ['average', 'poor', 'good', 'exceptional', 'good', 'average'],
    'student': ['tasnima', 'maliha', 'arshi', 'farzana', 'bushra', 'siza']
})

df
```

```python
new_df = df.replace(['poor', 'average', 'good', 'exceptional'], [1, 2, 3, 4])
new_df
```
